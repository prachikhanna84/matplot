```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_df=pd.read_csv(city_data_to_load)
ride_df=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

merged_df=pd.merge(ride_df,city_df,on="city",how="left")
# Display the data table for preview
merged_df.head()
# merged_df["driver_count"].unique()
```

## Bubble Plot of Ride Sharing Data

```python
# * Average Fare ($) Per City
# * Total Number of Rides Per City
# * Total Number of Drivers Per City
# * City Type (Urban, Suburban, Rural)

f, axs = plt.subplots(figsize=(6,4))

for x in merged_df["type"].unique():
# Obtain the x and y coordinates for each of the three city types
    type_df=merged_df.loc[merged_df["type"]==x]
    average_fare=type_df.groupby(["city"]).mean()
    average_fare=average_fare["fare"]

    total_ride=type_df.groupby(["city"]).count()
    total_ride=total_ride["ride_id"]
    
    driver_per_city=city_df.loc[city_df["type"]==x]
    driver_per_city=driver_per_city.groupby(["city"]).sum()
    driver_per_city=driver_per_city["driver_count"]
    
    # Build the scatter plots for each city types

    # Incorporate the other graph properties

    if x=="Urban":
        colorType="tomato"
    elif x=="Suburban":
        colorType="lightblue"
    else:
        colorType="yellow"
  
    plt.scatter(total_ride, average_fare, marker="o", facecolors=colorType, edgecolors="black",
                s=driver_per_city * 10
                , alpha=.75, label=x)
    plt.legend()


# Incorporate a text label regarding circle size

plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Rider Sharing Data (2016)")
plt.grid(True)
lgnd=plt.legend(title="City Types")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
plt.text(43, 34, "Circle Size Correlates with Driver Count Per city", fontsize=12)
plt.text(43, 36, "Note:", fontsize=12)

# Save Figure

plt.savefig("ridesSharing.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# * % of Total Fares by City Type
# * % of Total Rides by City Type
# * % of Total Drivers by City Type

total_fare=merged_df.groupby(["type"]).sum()["fare"]
# Calculate Type Percents


# Labels for the sections of our pie chart
labels = ["Rural","Suburban", "Urban"]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)


# Build Pie Chart
plt.pie(total_fare, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("farebyCity.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ride=merged_df.groupby(["type"]).count()["ride_id"]

# Build Pie Chart
plt.pie(total_ride, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("ridebyCity.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Ride Percents
driver_per_citytype=city_df.groupby(["type"]).sum()["driver_count"]

# Build Pie Chart
plt.pie(driver_per_citytype, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("driverbyCity.png")
```

```python
# Show Figure
plt.show()
```
